```python
import urllib
import urllib.request
from bs4 import BeautifulSoup
```

```python
def make_soup(theurl):
    thepage = urllib.request.urlopen(theurl)
    soup = BeautifulSoup(thepage, 'html.parser')
    return soup
```

```python
soup = make_soup("https://twitter.com/AndrewYNg")
```
